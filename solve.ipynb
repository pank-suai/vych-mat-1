## Задание 1

$$
   x^3 - \cos x = 0
$$

```kotlin
%useLatestDescriptors
%use lets-plot, dataframe
```

```kotlin
val func = { x: Double -> x * x * x - cos(x) }


letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) +  geomFunction(fn = func, color = "red")
```

### Метод половинного деления

```kotlin
// Точность

val epsilon = 1e-6
```

```kotlin

var a = 0.0
var b = 1.0

val data = mapOf(
    "x" to mutableListOf<Double>(),
    "y" to mutableListOf<Double>(),
    "a" to mutableListOf<Double>(),
    "b" to mutableListOf<Double>()
)

while (true) {
    val x = (a + b) / 2
    val y = func(x)

    data["x"]?.add(x)
    data["y"]?.add(y)
    data["a"]?.add(a)
    data["b"]?.add(b)

    if (abs(y) < epsilon) break

    if (func(a) * y < 0) {
        b = x
    } else {
        a = x
    }
}
```

```kotlin
val df = data.toDataFrame().addId("n")
df
```

```kotlin
letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) +  geomFunction(fn = func, color = "red") +
    geomPoint(df.toMap(), color = "blue", size = 1.8){
        x = "x"
        y = "y"
    }
```

### Метод хорд

```kotlin
a = 0.0
b = 1.0


// Очистим данные
data.forEach { _, list ->
    list.clear()
}
```

```kotlin
while (true) {
    val x = (a * func(b) - b * func(a)) / (func(b) - func(a))
    val y = func(x)

    data["x"]?.add(x)
    data["y"]?.add(y)
    data["a"]?.add(a)
    data["b"]?.add(b)

    if (abs(y) < epsilon) break

    if (func(a) * y < 0) {
        b = x
    } else {
        a = x
    }
}
```

```kotlin
val df = data.toDataFrame().addId("n")
df
```

```kotlin
letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) + geomFunction(fn = func, color = "red") +
        geomPoint(df.toMap(), color = "blue", size = 1.8){
            x = "x"
            y = "y"
        }
```

### Метод Ньютона

```kotlin
val dataNuton = mapOf(
    "x" to mutableListOf<Double>(),
    "y" to mutableListOf<Double>()
)

val dirrative = { x: Double -> 3 * x * x + sin(x) }
val dir2 = {x: Double -> 6 * x + cos(x) }
```

```kotlin
var x = 1.0
var y = func(x)

dataNuton["x"]?.add(x)
dataNuton["y"]?.add(y)


while (true) {
    x = x - func(x) / dirrative(x)
    y = func(x)

    dataNuton["x"]?.add(x)
    dataNuton["y"]?.add(y)

    if (abs(y) < epsilon) break
}
```

```kotlin
val df = dataNuton.toDataFrame().addId("n")
df
```

```kotlin
letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) + geomFunction(fn = func, color = "red") +
        geomPoint(df.toMap(), color = "blue", size = 1.8){
            x = "x"
            y = "y"
        }
```

### Ньютон (модифицированный)

```kotlin
dataNuton["x"]?.clear()
dataNuton["y"]?.clear()
```

```kotlin
var prevX = 1.0
var x = 0.866021
var y = func(x)

dataNuton["x"]?.add(x)
dataNuton["y"]?.add(y)


while (true) {
    val newX = x - (func(x) * (x - prevX)) / (func(x) - func(prevX))
    prevX = x
    x = newX
    y = func(x)

    dataNuton["x"]?.add(x)
    dataNuton["y"]?.add(y)

    if (abs(y) < epsilon) break
}
```

```kotlin
val df = dataNuton.toDataFrame().addId("n")
df
```

```kotlin
letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) + geomFunction(fn = func, color = "red") +
        geomPoint(df.toMap(), color = "blue", size = 1.8){
            x = "x"
            y = "y"
        }
```

### Метод простых итераций

```kotlin
val lambda = 0.3333

dataNuton["x"]?.clear()
dataNuton["y"]?.clear()
```

```kotlin
var x = 1.0
dataNuton["x"]?.add(x)
dataNuton["y"]?.add(func(x))
```

```kotlin
var i = 0

while (true){
    x = x - lambda * func(x)
    y = func(x)

    dataNuton["x"]?.add(x)
    dataNuton["y"]?.add(y)

    if (abs(y) < epsilon) break
    i++
    if (i > 100) break
}
```

```kotlin
val df = dataNuton.toDataFrame().addId("n")
df
```

```kotlin
letsPlot {  } + geomHLine(yintercept = 0) + geomVLine(xintercept = 0) + geomFunction(fn = func, color = "red") +
        geomPoint(df.toMap(), color = "blue", size = 1.8){
            x = "x"
            y = "y"
        }
```
